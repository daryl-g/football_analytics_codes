```python
import pandas as pd
```

```python
xg_data = pd.read_csv('A-League 2021-22 xG table.csv', delimiter = ';')
xg_data
```

```python
xg_table = pd.DataFrame()
teams = ['Adelaide United', 'Brisbane Roar', 'Central Coast Mariners', 'Macarthur', 'Melbourne City', 'Melbourne Victory',
         'Newcastle Jets', 'Perth Glory', 'Sydney', 'Wellington Phoenix', 'Western Sydney Wanderers', 'Western United']
team_info = {
    'Team name': '',
    'Matches played': 0,
    'Win': 0,
    'Draw': 0,
    'Lose': 0,
    'xG': 0,
    'Goals scored': 0,
    'xGA': 0,
    'Goals conceded': 0,
    'xPoints': 0,
    'Actual points': 0
}

for team in teams:
    
    matches_played = 0
    win = 0
    draw = 0
    lose = 0
    xG = 0
    xGA = 0
    goals_scored = 0
    goals_conceded = 0
    xPoints = 0
    total_points = 0
    
    for i in range(len(xg_data)):
        
        if (xg_data['Team'][i] == team):
            
            matches_played = matches_played + 1
            xG = xG + float(xg_data['xG for'][i])
            xGA = xGA + float(xg_data['xG against'][i])
            goals_scored = goals_scored + int(xg_data['Goals'][i])
            goals_conceded = goals_conceded + int(xg_data['Goals conceded'][i])
            
            if (xg_data['Goals'][i] > xg_data['Goals conceded'][i]):
                total_points = total_points + 3
                win = win + 1
            elif (xg_data['Goals'][i] == xg_data['Goals conceded'][i]):
                total_points = total_points + 1
                draw = draw + 1
            else:
                lose = lose + 1
                
            if (xg_data['xG difference'][i] >= 1.5):
                xPoints = xPoints + 2.7
            elif (1 <= xg_data['xG difference'][i]) and (xg_data['xG difference'][i] < 1.5):
                xPoints = xPoints + 2.3
            elif (0.5 <= xg_data['xG difference'][i]) and (xg_data['xG difference'][i] < 1):
                xPoints = xPoints + 2
            elif (0 <= xg_data['xG difference'][i]) and (xg_data['xG difference'][i] < 0.5):
                xPoints = xPoints + 1.5
            elif (-0.5 <= xg_data['xG difference'][i]) and (xg_data['xG difference'][i] < 0):
                xPoints = xPoints + 0.7
            elif (-1 <= xg_data['xG difference'][i]) and (xg_data['xG difference'][i] < -0.5):
                xPoints = xPoints + 0.5
            elif (-1.5 <= xg_data['xG difference'][i]) and (xg_data['xG difference'][i] < -1):
                xPoints = xPoints + 0.3
            else:
                xPoints = xPoints + 0.1
                
    team_info['Team name'] = team
    team_info['Matches played'] = matches_played
    team_info['Win'] = win
    team_info['Draw'] = draw
    team_info['Lose'] = lose
    team_info['xG'] = xG
    team_info['Goals scored'] = goals_scored
    team_info['xGA'] = xGA
    team_info['Goals conceded'] = goals_conceded
    team_info['xPoints'] = xPoints
    team_info['Actual points'] = total_points
    
    xg_table = xg_table.append(team_info, ignore_index = True)
    
xg_table = xg_table.sort_values('Actual points', ascending = False).reset_index()
xg_table = xg_table.drop('index', axis = 1)
xg_table.index = xg_table.index + 1

xg_table = xg_table.sort_values('xPoints', ascending = False).reset_index()
xg_table.rename({'index': 'Actual position'}, inplace = True, axis = 1)
xg_table.index = xg_table.index + 1
xg_table
```
