```python
import pandas as pd
import matplotlib as mpl
from numpy import mean
import matplotlib.pyplot as plt
mpl.rcParams['figure.dpi'] = 300
```

```python
npl_attacking_stats = pd.read_csv("Attacking data.csv", delimiter = ';')

npl_attacking_stats_backup = npl_attacking_stats.drop(['Teams'], axis = 1)
npl_attacking_stats_backup = npl_attacking_stats_backup.drop(14, axis = 0)

team_attacking = npl_attacking_stats.drop(["Shots", "Chances, % of conversion"], axis = 1)
team_attacking = npl_attacking_stats.drop(14, axis = 0)

teams = team_attacking['Teams']
```

```python
shots = npl_attacking_stats_backup['Shots']
chances_conversion = npl_attacking_stats_backup["Chances, % of conversion"]

shots_average = npl_attacking_stats['Shots'][14]
chances_conversion_average = npl_attacking_stats["Chances, % of conversion"][14]
```

```python
colors = []

for i in range(0, len(teams)):
    if (teams[i] == 'Olympic'):
        colors.append('white')
    elif (teams[i] == 'Logan Lightning'):
        colors.append('white')
    else:
        colors.append('white')
```

```python
min_y_value = shots[0]
max_x_value = 0
min_x_value = chances_conversion[0]
max_y_value = 0

for i in range(1, len(shots)):
    if (shots[i] < min_y_value):
        min_y_value = shots[i]

    if (chances_conversion[i] < min_x_value):
        min_x_value = chances_conversion[i]

for i in range(0, len(shots)):
    if (shots[i] > max_y_value):
        max_y_value = shots[i]
    
    if (chances_conversion[i] > max_x_value):
        max_x_value = chances_conversion[i]

olympic_y_value = shots[1]
olympic_x_value = chances_conversion[1]
opposition_y_value = shots[9]
opposition_x_value = chances_conversion[9]
```

```python
brisbane_strikers = plt.imread('Brisbane Strikers.png')
capalaba = plt.imread('Capalaba.png')
easts_suburbs = plt.imread("Eastern Suburbs.png")
gc_knights = plt.imread("GC Knights.png")
gc_united = plt.imread("GC United.png")
lions_fc = plt.imread("Lions.png")
logan_lightning = plt.imread("Logan Lightning.png")
magpies_crusaders = plt.imread("Magpies Crusaders.png")
moreton_bay = plt.imread("Moreton Bay.png")
olympic = plt.imread("Olympic.png")
pen_power = plt.imread("Peninsula Power.png")
redlands = plt.imread("Redlands.png")
roar_youth = plt.imread("Brisbane Roar.png")
sunny_coast_w = plt.imread("SC Wanderers.png")

npl_teams = [
    pen_power,
    lions_fc,
    olympic,
    roar_youth,
    moreton_bay,
    gc_knights,
    sunny_coast_w,
    gc_united,
    easts_suburbs,
    logan_lightning,
    capalaba,
    redlands,
    brisbane_strikers,
    magpies_crusaders
]
```

```python
fig, axs = plt.subplots(figsize = (8, 6))

for i in range(len(shots)):
        axs.scatter(chances_conversion, shots, s = 30, c = colors, linewidth = 1, alpha = 0)
        # axs.imshow(npl_teams[i], extent=[chances_conversion[i] - 1.5, chances_conversion[i], shots[i] - 1.5, shots[i]])

        if i == 2:
                axs.imshow(npl_teams[i], extent=[chances_conversion[i] - 1.3, chances_conversion[i], shots[i] - 1.5, shots[i]])
        else:
                axs.imshow(npl_teams[i], extent=[chances_conversion[i] - 1.5, chances_conversion[i], shots[i] - 1.5, shots[i]])

plt.axvline(chances_conversion_average, color = 'grey', linestyle = '--', alpha = 0.7)
plt.axhline(shots_average, color = 'grey', linestyle = '--', alpha = 0.7)

plt.annotate('Aggressive, Wasteful', xy=(min_x_value, max_y_value), xytext=(min_x_value - 1, max_y_value - 0.7), 
                color = 'Red', size = 10)
plt.annotate('Aggressive, Clinical', xy=(max_x_value, max_y_value), xytext=(max_x_value - 4.7, max_y_value - 0.7), 
                color = 'Red', size = 10)
plt.annotate('Passive, Wasteful', xy=(min_x_value, min_y_value), xytext=(min_x_value - 1, min_y_value - 1.2), 
                color = 'Green', size = 10)
plt.annotate('Passive, Clinical', xy=(min_x_value, max_y_value), xytext=(max_x_value - 3.9, min_y_value - 1.2), 
                color = 'Green', size = 10)

axs.set(title = 'Attacking efficiency (Stats are per game basis)\nNPL Queensland, This season', 
        xlabel = 'Chances conversion, %', ylabel = 'Shots (per game)')

fig.set_facecolor("#ffffff")
plt.tight_layout()
```
